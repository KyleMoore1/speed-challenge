```python
#imports
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import csv
import cv2
import seaborn as sns
from tqdm import tqdm
from fastai.vision import *
from PIL import Image
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
# constants
TRAIN_FRAMES = './data/IMG/train/'
TEST_FRAMES = './data/IMG/test/'
train_frames = 20400
test_frames = 10798
```

```python
# reading validation set into dataframe
train = pd.read_csv('train2.csv')
valid = train.loc[train['is_valid'] == True].reset_index().drop(['index', 'Unnamed: 0'], axis = 1)
valid.head()
```

```python
valid.shape
```

```python
def squared_error(x,y):
    return (x-y)**2
```

```python
def predict_velocities(df, model):
    predictions = []
    for filename in df['filename']:
        learn = load_learner('models', model)
        image = open_image('./data/IMG/' + filename)
        predictions.append(learn.predict(image)[0].data[0]) #get prediction
    return predictions
```

```python
predictions = predict_velocities(valid, 'model-2.38')
```

```python
def get_weighted_avg(predictions, beta):
    preds = copy(predictions)
    for i in range(1,2333):
        preds[i] = (beta) * preds[i-1] + (1-beta) * preds[i]
    return preds
```

```python
predictions[0]
```

```python
beta_values = np.arange(0, 1.1, 0.001)
errors = []
for beta in beta_values:
    y_true = valid['velocity']
    y_pred = get_weighted_avg(predictions, beta)
    errors.append(mean_squared_error(y_true, y_pred))
    
```

```python
errors
```

```python
best_index = errors.index(min(errors))
predictions_best = get_weighted_avg(predictions, beta_values[best_index])
valid['prediction'] = predictions_best
beta_values[best_index]
```

```python
min(errors)
```

```python
valid
```

```python
valid['error'] = squared_error(valid['velocity'], valid['prediction'])
```

```python
plt.figure(figsize = (20,10))
plt.plot(valid['frame_id'], valid['velocity'], 'r',  valid['frame_id'], valid['prediction'], 'xb', linewidth = 3)
```
